```python
# replace nan values in postal code with thiss : 05401
data['Postal Code'].fillna('05401', inplace=True)
```

```python
# 1. Understand Dataset
print("Number of rows:", len(data))
print("Number of columns:", len(data.columns))
print("Columns:", data.columns)
```

```python
# 2. Data Cleaning
print(data.isnull().sum())
data.dropna(inplace=True)
```

```python
# 3. Data Types
print(data.dtypes)
```

```python
# 4. Descriptive Statistics
print(data.describe())
```

```python
import matplotlib.pyplot as plt

# Assuming 'Sales' is a column in your dataset named 'data'
sales_data = data['Sales']

# Create the histogram
plt.hist(sales_data, bins=20, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.title('Histogram of Sales')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'Order Date' is a datetime object or can be converted to datetime
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Group sales by year
sales_by_year = data.groupby(data['Order Date'].dt.year)['Sales'].sum()

# Plotting
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Sales Trends Over 4 Years')
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Assuming 'Order Date' is a datetime object or can be converted to datetime
data['Order Date'] = pd.to_datetime(data['Order Date'])
data.set_index('Order Date', inplace=True)  # Set 'Order Date' as the index

# Perform seasonal decomposition
result = seasonal_decompose(data['Sales'], model='additive', period=12)  # Assuming seasonality is yearly

# Plot the decomposition
result.plot()
plt.show()
```

```python
import seaborn as sns

# Create a box plot for 'Sales'
sns.boxplot(x=data['Sales'])
plt.xlabel('Sales')
plt.title('Box Plot of Sales')
plt.show()
```

```python
# Create a bar chart for sales distribution by category
sales_by_category = data.groupby('Category')['Sales'].sum().sort_values(ascending=False)
plt.bar(sales_by_category.index, sales_by_category.values, color='skyblue')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.title('Sales Distribution by Category')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Group sales by month
monthly_sales = data.resample('M')['Sales'].sum()

# Plot monthly sales variation
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Variation')
plt.grid(True)
plt.show()
```

```python
correlation_matrix = data.corr()
print(correlation_matrix['Sales'])
```

```python
import matplotlib.pyplot as plt

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Sales'], data['Postal Code'])
plt.xlabel('Sales')
plt.ylabel('Postal Code')
plt.title('Scatter Plot of Sales vs. Postal Code')
plt.show()
```

```python
import seaborn as sns

# Select numeric columns for pair plot
numeric_columns = data.select_dtypes(include='number')

# Create pair plot
sns.pairplot(numeric_columns)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group data by category and sum sales
category_sales = data.groupby('Category')['Sales'].sum()

# Create line chart
category_sales.plot(kind='line', marker='o')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.title('Category-wise Sales Trends')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Calculate correlation matrix
correlation_matrix = data.corr()

# Create heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
region_sales = data.groupby('Region')['Sales'].sum().reset_index()

# Create stacked bar chart
region_sales.plot(kind='bar', x='Region', y='Sales', stacked=True)
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.title('Sales by Region')
plt.show()
```

```python
customer_segment_sales = data.groupby('Segment')['Sales'].sum()

# Create bar chart
customer_segment_sales.plot(kind='bar')
plt.xlabel('Customer Segment')
plt.ylabel('Total Sales')
plt.title('Sales by Customer Segment')
plt.show()
```

```python
product_sales = data.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

# Create bar chart
product_sales.plot(kind='bar')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.title('Top 10 Products by Sales')
plt.xticks(rotation=45)
plt.show()
```
